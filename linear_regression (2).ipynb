```python
import pandas as pd
```

```python
BostonHousing = pd.read_csv(r"/content/sample_data/BostonHousing.csv")
```

```python
BostonHousing.head()
```

CRIM: This column represents the per capita crime rate by town. It provides information about the crime rate in the area where the house is located.

ZN: This column represents the proportion of residential land zoned for lots larger than 25,000 sq.ft. It provides information about the amount of land allocated for residential purposes.

INDUS: This column represents the proportion of non-retail business acres per town. It gives insights into the extent of industrial and commercial land use in the area.

CHAS: This column represents the Charles River dummy variable. It takes a value of 1 if the tract bounds the river and 0 otherwise. This variable allows studying the impact of properties near the Charles River on housing prices.

NOX: This column represents the nitric oxides concentration (parts per 10 million). It gives an indication of air pollution levels in the area.

RM: This column represents the average number of rooms per dwelling. It provides information about the size of the houses in terms of the number of rooms.

AGE: This column represents the proportion of owner-occupied units built before 1940. It gives insights into the age of the houses in the area.

DIS: This column represents the weighted distances to five Boston employment centers. It gives an indication of the accessibility to employment centers from the houses in the area.

RAD: This column represents the index of accessibility to radial highways. It provides information about the accessibility to highways, which is essential for commuting.

TAX: This column represents the full-value property tax rate per $10,000. It gives insights into the property tax rates in different areas.

PTRATIO: This column represents the pupil-teacher ratio by town. It provides information about the average number of students per teacher in the local schools.

B: This column represents the 1000(Bk - 0.63)^2 where Bk is the proportion of Black people by town. It gives insights into the racial demographics of the area.

LSTAT: This column represents the percentage of lower status of the population. It gives an indication of the socio-economic status of the residents in the area.

MEDV: This column represents the median value of owner-occupied homes in $1000s. It is the target variable in regression analysis, representing the predicted housing price based on the given features.

```python
BostonHousing.describe()
```

```python
BostonHousing.isnull().sum()
```

```python
Y = BostonHousing.medv
Y
```

```python
X = BostonHousing.drop(['medv'], axis=1)
X
```

DATA SPLIT

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

LINEAR_REGRESSION_MODEL

```python
from sklearn import linear_model
from sklearn.metrics import mean_squared_error, r2_score
```

```python
model = linear_model.LinearRegression()
```

```python
model.fit(X_train, Y_train)
```

```python
Y_pred = model.predict(X_test)
```

PREDICTION RESULT

```python
print('Coefficients:', model.coef_)
print('Intercept:', model.intercept_)
print('Mean squared error (MSE): %.2f'
      % mean_squared_error(Y_test, Y_pred))
print('Coefficient of determination (R^2): %.2f'
      % r2_score(Y_test, Y_pred))
```
